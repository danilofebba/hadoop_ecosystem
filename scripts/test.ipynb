```python
schema = {
    "type": "record",
    "name": "test_data",
    "fields": [
        {
            "name": "id",
            "type": {
                "type": "string",
                "logicalType": "uuid"
            }
        },
        {
            "name": "code",
            "type": "long"
        },
        {
            "name": "option",
            "type": "string"
        },
        {
            "name": "description",
            "type": "string"
        },
        {
            "name": "value",
            "type": "double"
        },
        {
            "name": "rate",
            "type": "double"
        },
        {
            "name": "start_date",
            "type": {
                "type": "int",
                "logicalType": "date"
            }
        },
        {
            "name": "created_at",
            "type": {
                "type": "long",
                "logicalType": "local-timestamp-micros"
            }
        },
        {
            "name": "updated_at",
            "type": {
                "type": "long",
                "logicalType": "local-timestamp-micros"
            }
        },
        {
            "name": "status",
            "type": "boolean"
        },
        {
            "name": "detail",
            "type": {
                "name": "test_data.details",
                "type": "record",
                "fields": [
                    {
                        "name": "detail_id",
                        "type": "long"
                    },
                    {
                        "name": "description",
                        "type": "string"
                    }
                ]
            }
        }
    ]
}
schema_parsed = avro.schema.Parse(json.dumps(schema))
```
